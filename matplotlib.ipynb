**Numpy Arrays**

```python
import numpy as np
x=np.array([1,2])
y=np.array([8,5])
plt.scatter(x,y,color=['r','b'])
plt.show()
```

**2D Vectors**

```python
#2D vectors
u=np.array([1,2])
v=np.array([2,5])

x,y=zip(u,v)
plt.scatter(x,y,color=['r','b'])
plt.show()
```

**Since vectors are usually represented with arrows, lets create a quick convenience function to draw arrows for the vectors u
 and v
:**

```python
def plot_vector2d(vector2d,origin=[0,0],**options):
  return plt.arrow(origin[0],origin[1],vector2d[0],vector2d[1],head_width=0.2,head_length=0.3,length_includes_head=True,**options)

u = np.array([1, 2])
v = np.array([3, 4])

x, y = zip(u, v)
plt.scatter(x, y, color=["r","b"])
plot_vector2d(u, color="r")
plot_vector2d(v, color="b")
plt.axis([0, 9, 0, 6])
plt.show()
```

**Subplots**

```python
x = np.linspace(-2,2,100)


plt.subplot(2,2,1)
plt.plot(x,x)

plt.subplot(2,2,2)
plt.plot(x,x**2)

plt.subplot(2, 2, 3)
plt.plot(x,x*3)

plt.subplot(2, 2, 4)
plt.plot(x,x**4)
```

**Histograms**

```python
data=[1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(2, 1, 1)
plt.hist(data, bins = 10, rwidth=0.8)


plt.subplot(2, 1, 2)
plt.hist(data, bins=5, rwidth=0.95)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
